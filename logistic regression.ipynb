```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
DF= pd.read_csv('ChurnData.csv')
DF.head()
DF.describe()
```

```python
X=DF[['age','income','employ']]
Y=DF[['churn']]
```

```python
from sklearn import preprocessing
x=preprocessing.StandardScaler().fit(X).transform(X)
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y= train_test_split(x,Y ,test_size=0.2 ,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
log= LogisticRegression(C=0.01, solver='liblinear').fit(train_x, train_y)
```

```python
y_hat=log.predict(test_x)
```

```python
from sklearn.metrics import confusion_matrix
print(f"{confusion_matrix(y_hat, test_y)}")
```

```python
from sklearn.metrics import jaccard_score
print(f"{jaccard_score(y_hat, test_y, pos_label=0)}")
```

```python
from sklearn.metrics import r2_score
print(f"{r2_score(y_hat, test_y)}")
```

```python
from sklearn.metrics import log_loss
print(f"{log_loss(test_y, y_hat)}")
```
