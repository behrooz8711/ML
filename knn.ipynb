```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pylab as pl
%matplotlib inline
```

```python
df= pd.read_csv('teleCust1000t.csv')
```

```python
df.describe()
```

```python
df.head()
```

```python
x=df[['region','age','marital','income','ed','reside']]
y=df[['custcat']]
x,y
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
x=preprocessing.StandardScaler().fit(x).transform(x.astype(float))
```

```python
train_x, test_x,train_y,test_y= train_test_split(x,y , test_size=.2, random_state=4)
```

```python
train_x, train_y
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k=4
knn= KNeighborsClassifier(n_neighbors=k).fit(train_x,train_y)
```

```python
y_hat= knn.predict(test_x)
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score
print(metrics.accuracy_score(test_y, y_hat))
print(r2_score( y_hat, test_y,))
```
