```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
DF= pd.read_csv('cell_samples.csv')
DF.head()
#DF.describe()
```

```python
DF.dtypes
```

```python
df = DF[pd.to_numeric(DF['BareNuc'], errors='coerce').notnull()]
df['BareNuc'] = df['BareNuc'].astype('int')  # Convert BareNuc to int

# Now you can access all columns:
x = np.asarray(df[['Clump','UnifSize','UnifShape','MargAdh','SingEpiSize','BareNuc','BlandChrom','NormNucl','Mit']])
```

```python
y= np.asarray(df[['Class']])
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y= train_test_split(x,y ,test_size=0.2 ,random_state=0)
```

```python
from sklearn import svm
svm= svm.SVC(kernel='rbf').fit(train_x, train_y)
```

```python
y_hat=svm.predict(test_x)
```

```python
from sklearn.metrics import f1_score, jaccard_score
print(f"{f1_score(y_hat,test_y, average='weighted')},{jaccard_score(y_hat,test_y, pos_label=2)}")
```
